### (3.2 -- 1%) When it comes to comparing images, what does SSIM capture that PSNR does not?

Answer: SSIM captures how well the image is preserved, including features like structure, luminance, and contrast. PSNR looks at the error in the image (average squared error), rather than the image's qualities. This means that two images can have the same PSNR but look very different, while SSIM can help to show how the images are similar.
