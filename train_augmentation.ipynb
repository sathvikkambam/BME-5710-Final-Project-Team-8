Import libraries

```python
import numpy as np
from PIL import Image
import os
```

Load training images and makes parameters

```python
files = sorted([f for f in os.listdir("train") if not f.startswith('.')])

old_folder = "train"
new_folder = "train_augmented"
imag_size = 256
```

TESTING Deletes all images in new folder

```python
if os.path.exists(new_folder):
    filesRM = sorted([f for f in os.listdir(new_folder) if not f.startswith('.')])
    for f in filesRM:
        os.remove(os.path.join(new_folder, f))
```

Created new folder: "train_augmented"

```python
if not os.path.exists(new_folder):
    os.makedirs(new_folder, exist_ok=True)
    print(f"Created directory: {new_folder}")
else:
    print(f"Directory already exists: {new_folder}")

```

Iterates through each tiff training file and loads it

```python
images = []
for fname in files:
    with Image.open(os.path.join(old_folder, fname)) as im:
        img = np.array(im)
    if img.dtype == np.uint16:                         # 16-bit → divide by 65535
        img = img.astype(np.float32) / 65535.0
    elif img.dtype == np.uint8:                        # 8-bit → divide by 255
        img = img.astype(np.float32) / 255.0
    else:                                                # already float → use as-is
        img = img.astype(np.float32)
    assert img.shape == (imag_size, imag_size), (
        f"Unexpected shape {img.shape} for {fname}; expected ({imag_size}, {imag_size})"
    )
    images.append(img)

print(f"Loaded {len(images)} images from '{old_folder}'")
```

Iterates through each image and transforms it while saving to "train_augmented"

```python
#Augmentations include: normal, horizontal flip, vertical flip, and 90 degree rotation
for i in range(len(images)):
    img = images[i]
    fname = files[i].split('.')[0]      #Isolates image and filename without extension
    extension = files[i].split('.')[1]  #Isolates extension
    Image.fromarray(img).save(os.path.join(new_folder, fname + "_normal." + extension))      #Saves normal image
    Image.fromarray(np.rot90(img)).save(os.path.join(new_folder, fname + "_rot90CCW." + extension))  #Saves 90 degree rotated counterclock wise image
    Image.fromarray(np.rot90(img, k=2)).save(os.path.join(new_folder, fname + "_rot180." + extension))  #Saves 180 degree rotated image
    Image.fromarray(np.rot90(img, k=3)).save(os.path.join(new_folder, fname + "_rot90CW." + extension))  #Saves 90 degree rotated clockwise image
    Image.fromarray(np.flipud(img)).save(os.path.join(new_folder, fname + "_vflip." + extension))  #Saves vertically flipped image
    Image.fromarray(np.fliplr(img)).save(os.path.join(new_folder, fname + "_hflip." + extension))  #Saves horizontally flipped image
    Image.fromarray(np.rot90(np.fliplr(img))).save(os.path.join(new_folder, fname + "_hflip_rot90CCW." + extension))  #Saves horizontally flipped + 90 degree rotated counterclock wise image
    Image.fromarray(np.rot90(np.fliplr(img), k=3)).save(os.path.join(new_folder, fname + "_hflip_rot90CW." + extension))  #Saves horizontally flipped + 90 degree rotated clockwise image
```

Counts number of files in new directory

```python
files_new = sorted([f for f in os.listdir(new_folder) if not f.startswith('.')])
print(f"Saved {len(files_new)} augmented images to '{new_folder}'")
```
